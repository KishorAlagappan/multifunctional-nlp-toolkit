```python
import pandas as pd
from deep_translator import GoogleTranslator
```

Load Dataset

```python
dataframe = pd.read_csv('English.csv')
dataframe
```

Sample Text
  *  Here, our dataset has '175620' english words/ sentences, so now I am doing sampling.

```python
sample_dataframe = dataframe.iloc[::1756]

sample_dataframe.reset_index(drop=True, inplace=True)

french_list = []
german_list = []
spanish_list = []
```

```python
sample_dataframe
```

Translator object for each language
  * English (Existing Language)
  * German
  * French
  * Spanish

```python
french_translator = GoogleTranslator(source='english', target='french')
german_translator = GoogleTranslator(source='english', target='german')
spanish_translator = GoogleTranslator(source='english', target='spanish')
```

```python
for i in (sample_dataframe.iloc[:, 0]):

    french_list.append(french_translator.translate(i))
    german_list.append(german_translator.translate(i))
    spanish_list.append(spanish_translator.translate(i))
```

```python
full_dataframe =  pd.DataFrame({'French':french_list, 'German':german_list, 'Spanish':spanish_list})
```

```python
translated_dataframe = pd.concat([sample_dataframe.iloc[:, 0], full_dataframe], axis=1)
```
